```python
import numpy as np
```

```python
dataset = np.load('2X2_pixels_inpainting.npy')
```

```python
# checking loading
images    = dataset[0]
gt_images = dataset[1] 

nSamples, ndim = gt_images.shape
print nSamples
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for SampleID in np.random.randint(nSamples,size=3):
    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(np.reshape(gt_images[SampleID],(28,28)), interpolation='None',vmin=-1, vmax=1)
    plt.title("GT image")
    plt.subplot(1,2,2)
    plt.imshow(np.reshape(images[SampleID],(28,28)), interpolation='None',vmin=-1, vmax=1)
    plt.title("Cropped image")
    #plt.savefig("sample_"+str(SampleID)+"_2x2_pixels_inpainting.png")
```
